## C1: Fine-tuning a Small LLM (15 pts)

Load IMDB, tokenize with DistilBERT tokenizer, fine-tune for 5 epochs.
Report training loss and training accuracy per epoch.

```python
# W&B run for C1 baseline
run = wandb.init(
    project="hpml-hw2-llm",
    name="c1-baseline",
    config={
        "model_name": "distilbert-base-uncased",
        "max_len": 256,
        "batch_size": 32,
        "lr": 1e-4,
        "optimizer": "AdamW",
        "num_workers": 2,
        "epochs": 5,
        "compile_mode": False,
    },
)
```

### Load and Tokenize IMDB Dataset

```python
dataset = load_dataset("imdb")
print(f"Train: {len(dataset['train']):,} examples")
print(f"Test:  {len(dataset['test']):,} examples")
print(f"\nSample review (first 200 chars):\n{dataset['train'][0]['text'][:200]}...")
print(f"Label: {dataset['train'][0]['label']}")
```

```python
tokenizer = DistilBertTokenizerFast.from_pretrained("distilbert-base-uncased")

def tokenize_fn(batch):
    return tokenizer(batch["text"], truncation=True, max_length=256)

tokenized = dataset.map(tokenize_fn, batched=True, batch_size=1000)
tokenized.set_format("torch", columns=["input_ids", "attention_mask", "label"])

# Show token length distribution
sample_lens = [len(tokenized["train"][i]["input_ids"]) for i in range(100)]
print(f"Token lengths (first 100): min={min(sample_lens)}, max={max(sample_lens)}, avg={sum(sample_lens)/len(sample_lens):.0f}")
```

### DataLoaders with Dynamic Padding

```python
def dynamic_padding(batch):
    """Dynamic padding: pad to max length within the batch, not global max."""
    input_ids = [item["input_ids"] for item in batch]
    attention_mask = [item["attention_mask"] for item in batch]
    labels = torch.stack([item["label"] for item in batch])

    input_ids = torch.nn.utils.rnn.pad_sequence(
        input_ids, batch_first=True, padding_value=tokenizer.pad_token_id
    )
    attention_mask = torch.nn.utils.rnn.pad_sequence(
        attention_mask, batch_first=True, padding_value=0
    )

    return {"input_ids": input_ids, "attention_mask": attention_mask, "labels": labels}


BATCH_SIZE = 32
NUM_WORKERS = 2

train_loader = DataLoader(
    tokenized["train"],
    batch_size=BATCH_SIZE,
    shuffle=True,
    num_workers=NUM_WORKERS,
    collate_fn=dynamic_padding,
    pin_memory=torch.cuda.is_available(),
)
test_loader = DataLoader(
    tokenized["test"],
    batch_size=BATCH_SIZE,
    shuffle=False,
    num_workers=NUM_WORKERS,
    collate_fn=dynamic_padding,
    pin_memory=torch.cuda.is_available(),
)

print(f"Train batches: {len(train_loader)}, Test batches: {len(test_loader)}")
```

### Load Model

```python
model = DistilBertForSequenceClassification.from_pretrained(
    "distilbert-base-uncased", num_labels=2
)
model = model.to(device)
optimizer = AdamW(model.parameters(), lr=1e-4)

total_params = sum(p.numel() for p in model.parameters())
trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
print(f"Total parameters:     {total_params:,}")
print(f"Trainable parameters: {trainable_params:,}")
```

### Training and Evaluation Functions

```python
def train_one_epoch(model, dataloader, optimizer, device):
    model.train()
    total_loss = 0.0
    correct = 0
    total = 0

    for batch in tqdm(dataloader, desc="Training"):
        input_ids = batch["input_ids"].to(device)
        attention_mask = batch["attention_mask"].to(device)
        labels = batch["labels"].to(device)

        optimizer.zero_grad()
        outputs = model(input_ids=input_ids, attention_mask=attention_mask, labels=labels)
        loss = outputs.loss
        loss.backward()
        optimizer.step()

        total_loss += loss.item() * labels.size(0)
        preds = outputs.logits.argmax(dim=-1)
        correct += (preds == labels).sum().item()
        total += labels.size(0)

    return total_loss / total, correct / total


@torch.no_grad()
def evaluate(model, dataloader, device):
    model.eval()
    correct = 0
    total = 0

    for batch in tqdm(dataloader, desc="Evaluating"):
        input_ids = batch["input_ids"].to(device)
        attention_mask = batch["attention_mask"].to(device)
        labels = batch["labels"].to(device)

        outputs = model(input_ids=input_ids, attention_mask=attention_mask)
        preds = outputs.logits.argmax(dim=-1)
        correct += (preds == labels).sum().item()
        total += labels.size(0)

    return correct / total
```

### Train for 5 Epochs

```python
NUM_EPOCHS = 5

for epoch in range(NUM_EPOCHS):
    t0 = time.perf_counter()
    train_loss, train_acc = train_one_epoch(model, train_loader, optimizer, device)
    test_acc = evaluate(model, test_loader, device)
    elapsed = time.perf_counter() - t0

    wandb.log({
        "epoch": epoch + 1,
        "train/loss": train_loss,
        "train/acc": train_acc,
        "test/acc": test_acc,
    })

    print(
        f"Epoch {epoch+1}/{NUM_EPOCHS} | "
        f"Loss: {train_loss:.4f} | "
        f"Train Acc: {train_acc:.4f} | "
        f"Test Acc: {test_acc:.4f} | "
        f"Time: {elapsed:.1f}s"
    )
```

## C2: Baseline Timing (10 pts)

For each epoch, measure and report:
- a) Data loading time
- b) Training computation time (forward + backward + optimizer)
- c) Total epoch time

Uses `torch.cuda.synchronize()` for accurate GPU timing.

```python
def train_one_epoch_timed(model, dataloader, optimizer, device, epoch=None, num_epochs=None):
    """Training loop with separate timing for data loading vs compute."""
    model.train()
    total_loss = 0.0
    correct = 0
    total = 0
    data_time = 0.0
    compute_time = 0.0

    desc = f"Epoch {epoch}/{num_epochs}" if epoch else "Training"
    t_end = time.perf_counter()

    for batch in tqdm(dataloader, desc=desc):
        # --- Data loading time ---
        if device.type == "cuda":
            torch.cuda.synchronize()
        t_data = time.perf_counter()
        data_time += t_data - t_end

        # --- Compute time (forward + backward + optimizer) ---
        input_ids = batch["input_ids"].to(device)
        attention_mask = batch["attention_mask"].to(device)
        labels = batch["labels"].to(device)

        optimizer.zero_grad()
        outputs = model(input_ids=input_ids, attention_mask=attention_mask, labels=labels)
        loss = outputs.loss
        loss.backward()
        optimizer.step()

        if device.type == "cuda":
            torch.cuda.synchronize()
        t_compute = time.perf_counter()
        compute_time += t_compute - t_data

        total_loss += loss.item() * labels.size(0)
        preds = outputs.logits.argmax(dim=-1)
        correct += (preds == labels).sum().item()
        total += labels.size(0)

        t_end = time.perf_counter()

    return total_loss / total, correct / total, data_time, compute_time
```

```python
# Re-initialize model and optimizer for a clean timed run
model = DistilBertForSequenceClassification.from_pretrained(
    "distilbert-base-uncased", num_labels=2
)
model = model.to(device)
optimizer = AdamW(model.parameters(), lr=1e-4)

run = wandb.init(
    project="hpml-hw2-llm",
    name="c2-baseline-timed",
    config={
        "model_name": "distilbert-base-uncased",
        "max_len": 256,
        "batch_size": 32,
        "lr": 1e-4,
        "optimizer": "AdamW",
        "num_workers": 2,
        "epochs": 5,
        "compile_mode": False,
    },
)

NUM_EPOCHS = 5

for epoch in range(NUM_EPOCHS):
    epoch_start = time.perf_counter()

    train_loss, train_acc, data_time, compute_time = train_one_epoch_timed(
        model, train_loader, optimizer, device, epoch=epoch+1, num_epochs=NUM_EPOCHS
    )
    test_acc = evaluate(model, test_loader, device)

    if device.type == "cuda":
        torch.cuda.synchronize()
    epoch_time = time.perf_counter() - epoch_start

    wandb.log({
        "epoch": epoch + 1,
        "train/loss": train_loss,
        "train/acc": train_acc,
        "test/acc": test_acc,
        "time/data_loading": data_time,
        "time/compute": compute_time,
        "time/epoch": epoch_time,
    })

    print(
        f"Epoch {epoch+1}/{NUM_EPOCHS} | "
        f"Loss: {train_loss:.4f} | "
        f"Train Acc: {train_acc:.4f} | "
        f"Test Acc: {test_acc:.4f}\n"
        f"  Data Loading: {data_time:.2f}s | "
        f"Compute: {compute_time:.2f}s | "
        f"Epoch Total: {epoch_time:.2f}s"
    )
```

```python
wandb.finish()
```

## C3: DataLoader Performance (10 pts)

Vary `num_workers` in {0, 2, 4, 8}. For each setting, report total data loading time and plot epoch time vs. num_workers. Identify the optimal number of workers.

```python
import matplotlib.pyplot as plt

NUM_WORKERS_LIST = [0, 2, 4, 8]
NUM_EPOCHS = 5
c3_results = []

for nw in NUM_WORKERS_LIST:
    print(f"\n{'='*60}")
    print(f"Training with num_workers={nw}")
    print(f"{'='*60}")

    # Create DataLoaders with this num_workers
    nw_train_loader = DataLoader(
        tokenized["train"],
        batch_size=32,
        shuffle=True,
        num_workers=nw,
        collate_fn=dynamic_padding,
        pin_memory=torch.cuda.is_available(),
    )
    nw_test_loader = DataLoader(
        tokenized["test"],
        batch_size=32,
        shuffle=False,
        num_workers=nw,
        collate_fn=dynamic_padding,
        pin_memory=torch.cuda.is_available(),
    )

    # Fresh model for each config
    model = DistilBertForSequenceClassification.from_pretrained(
        "distilbert-base-uncased", num_labels=2
    )
    model = model.to(device)
    opt = AdamW(model.parameters(), lr=1e-4)

    run = wandb.init(
        project="hpml-hw2-llm",
        name=f"c3-nw{nw}",
        config={
            "model_name": "distilbert-base-uncased",
            "max_len": 256,
            "batch_size": 32,
            "lr": 1e-4,
            "optimizer": "AdamW",
            "num_workers": nw,
            "epochs": NUM_EPOCHS,
            "compile_mode": False,
        },
    )

    total_data_time = 0.0
    epoch_times = []

    for epoch in range(NUM_EPOCHS):
        epoch_start = time.perf_counter()

        train_loss, train_acc, data_time, compute_time = train_one_epoch_timed(
            model, nw_train_loader, opt, device, epoch=epoch+1, num_epochs=NUM_EPOCHS
        )
        test_acc = evaluate(model, nw_test_loader, device)

        if device.type == "cuda":
            torch.cuda.synchronize()
        epoch_time = time.perf_counter() - epoch_start

        total_data_time += data_time
        epoch_times.append(epoch_time)

        wandb.log({
            "epoch": epoch + 1,
            "train/loss": train_loss,
            "train/acc": train_acc,
            "test/acc": test_acc,
            "time/data_loading": data_time,
            "time/compute": compute_time,
            "time/epoch": epoch_time,
        })

        print(
            f"Epoch {epoch+1}/{NUM_EPOCHS} | "
            f"Loss: {train_loss:.4f} | "
            f"Train Acc: {train_acc:.4f} | "
            f"Test Acc: {test_acc:.4f}\n"
            f"  Data: {data_time:.2f}s | Compute: {compute_time:.2f}s | Total: {epoch_time:.2f}s"
        )

    avg_epoch_time = sum(epoch_times) / len(epoch_times)
    c3_results.append({
        "num_workers": nw,
        "total_data_time": total_data_time,
        "avg_epoch_time": avg_epoch_time,
        "final_test_acc": test_acc,
    })

    wandb.finish()
```

```python
# Plot: Epoch Time vs num_workers
nw_vals = [r["num_workers"] for r in c3_results]
avg_times = [r["avg_epoch_time"] for r in c3_results]
data_times = [r["total_data_time"] for r in c3_results]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

ax1.bar(range(len(nw_vals)), avg_times, tick_label=[str(n) for n in nw_vals])
ax1.set_xlabel("num_workers")
ax1.set_ylabel("Avg Epoch Time (s)")
ax1.set_title("Epoch Time vs. num_workers")
for i, v in enumerate(avg_times):
    ax1.text(i, v + 0.5, f"{v:.1f}s", ha="center")

ax2.bar(range(len(nw_vals)), data_times, tick_label=[str(n) for n in nw_vals])
ax2.set_xlabel("num_workers")
ax2.set_ylabel("Total Data Loading Time (s)")
ax2.set_title("Total Data Loading Time vs. num_workers")
for i, v in enumerate(data_times):
    ax2.text(i, v + 0.5, f"{v:.1f}s", ha="center")

plt.tight_layout()
plt.show()

# Identify optimal
best = min(c3_results, key=lambda r: r["avg_epoch_time"])
print(f"\nOptimal num_workers = {best['num_workers']} (avg epoch time: {best['avg_epoch_time']:.2f}s)")
```

```python
# W&B Table comparing num_workers configurations
run = wandb.init(project="hpml-hw2-llm", name="c3-summary")

columns = ["num_workers", "total_data_time (s)", "avg_epoch_time (s)", "final_test_acc"]
table = wandb.Table(columns=columns)
for r in c3_results:
    table.add_data(
        r["num_workers"],
        round(r["total_data_time"], 2),
        round(r["avg_epoch_time"], 2),
        round(r["final_test_acc"], 4),
    )

wandb.log({"c3_num_workers_comparison": table})
wandb.finish()
```

## C4: PyTorch Profiler (15 pts)

Use `torch.profiler` with TensorBoard to generate profiler traces.
Profile 1 epoch with `num_workers=1` and the optimal value from C3.

Required deliverables:
- TensorBoard Profiler screenshots (Overview tab) for both configs
- Comparison table of time breakdown by category
- 2-3 paragraph analysis

```python
from torch.profiler import profile, record_function, ProfilerActivity, schedule, tensorboard_trace_handler

OPTIMAL_NW = best["num_workers"]  # from C3 results
PROFILE_CONFIGS = [1, OPTIMAL_NW]

for nw in PROFILE_CONFIGS:
    print(f"\n{'='*60}")
    print(f"Profiling with num_workers={nw}")
    print(f"{'='*60}")

    log_dir = f"./profiler_logs/nw{nw}"

    prof_train_loader = DataLoader(
        tokenized["train"],
        batch_size=32,
        shuffle=True,
        num_workers=nw,
        collate_fn=dynamic_padding,
        pin_memory=torch.cuda.is_available(),
    )

    # Fresh model
    model = DistilBertForSequenceClassification.from_pretrained(
        "distilbert-base-uncased", num_labels=2
    )
    model = model.to(device)
    opt = AdamW(model.parameters(), lr=1e-4)
    model.train()

    # Profile schedule: skip first 5 batches (warmup), then profile 15 batches
    prof_schedule = schedule(
        skip_first=5,
        wait=1,
        warmup=2,
        active=6,
        repeat=2,
    )

    with profile(
        activities=[ProfilerActivity.CPU, ProfilerActivity.CUDA],
        schedule=prof_schedule,
        on_trace_ready=tensorboard_trace_handler(log_dir),
        record_shapes=True,
        profile_memory=True,
        with_stack=True,
    ) as prof:
        for step, batch in enumerate(tqdm(prof_train_loader, desc=f"Profiling nw={nw}")):
            input_ids = batch["input_ids"].to(device)
            attention_mask = batch["attention_mask"].to(device)
            labels = batch["labels"].to(device)

            opt.zero_grad()
            with record_function("forward"):
                outputs = model(input_ids=input_ids, attention_mask=attention_mask, labels=labels)
            loss = outputs.loss
            with record_function("backward"):
                loss.backward()
            with record_function("optimizer_step"):
                opt.step()

            prof.step()

            # Stop after enough profiled steps
            if step >= 50:
                break

    print(f"Profiler traces saved to {log_dir}/")
    print(f"View with: tensorboard --logdir={log_dir}")
```

## C5: Hyperparameter Sensitivity (15 pts)

Sweep batch size {16, 32, 64} × learning rate {5e-5, 1e-4, 5e-4} = 9 configurations.
Report final training loss/accuracy, test accuracy, and total training time.
Discuss trade-offs among speed, accuracy, and stability.

```python
BATCH_SIZES = [16, 32, 64]
LEARNING_RATES = [5e-5, 1e-4, 5e-4]
NUM_EPOCHS = 5
c5_results = []

for bs in BATCH_SIZES:
    for lr in LEARNING_RATES:
        run_name = f"c5-bs{bs}_lr{lr}"
        print(f"\n{'='*60}")
        print(f"Training: batch_size={bs}, lr={lr}")
        print(f"{'='*60}")

        sweep_train_loader = DataLoader(
            tokenized["train"],
            batch_size=bs,
            shuffle=True,
            num_workers=OPTIMAL_NW,
            collate_fn=dynamic_padding,
            pin_memory=torch.cuda.is_available(),
        )
        sweep_test_loader = DataLoader(
            tokenized["test"],
            batch_size=bs,
            shuffle=False,
            num_workers=OPTIMAL_NW,
            collate_fn=dynamic_padding,
            pin_memory=torch.cuda.is_available(),
        )

        model = DistilBertForSequenceClassification.from_pretrained(
            "distilbert-base-uncased", num_labels=2
        )
        model = model.to(device)
        opt = AdamW(model.parameters(), lr=lr)

        run = wandb.init(
            project="hpml-hw2-llm",
            name=run_name,
            config={
                "model_name": "distilbert-base-uncased",
                "max_len": 256,
                "batch_size": bs,
                "lr": lr,
                "optimizer": "AdamW",
                "num_workers": OPTIMAL_NW,
                "epochs": NUM_EPOCHS,
                "compile_mode": False,
            },
        )

        total_train_time = 0.0

        for epoch in range(NUM_EPOCHS):
            epoch_start = time.perf_counter()

            train_loss, train_acc, data_time, compute_time = train_one_epoch_timed(
                model, sweep_train_loader, opt, device, epoch=epoch+1, num_epochs=NUM_EPOCHS
            )
            test_acc = evaluate(model, sweep_test_loader, device)

            if device.type == "cuda":
                torch.cuda.synchronize()
            epoch_time = time.perf_counter() - epoch_start
            total_train_time += epoch_time

            wandb.log({
                "epoch": epoch + 1,
                "train/loss": train_loss,
                "train/acc": train_acc,
                "test/acc": test_acc,
                "time/epoch": epoch_time,
            })

            print(
                f"Epoch {epoch+1}/{NUM_EPOCHS} | "
                f"Loss: {train_loss:.4f} | "
                f"Train Acc: {train_acc:.4f} | "
                f"Test Acc: {test_acc:.4f} | "
                f"Time: {epoch_time:.1f}s"
            )

        c5_results.append({
            "batch_size": bs,
            "lr": lr,
            "final_train_loss": train_loss,
            "final_train_acc": train_acc,
            "final_test_acc": test_acc,
            "total_time": total_train_time,
        })

        wandb.finish()
```

```python
# Print results table
print(f"{'BS':>4} | {'LR':>8} | {'Train Loss':>10} | {'Train Acc':>9} | {'Test Acc':>8} | {'Time (s)':>8}")
print("-" * 65)
for r in c5_results:
    print(
        f"{r['batch_size']:>4} | {r['lr']:>8.0e} | "
        f"{r['final_train_loss']:>10.4f} | {r['final_train_acc']:>9.4f} | "
        f"{r['final_test_acc']:>8.4f} | {r['total_time']:>8.1f}"
    )

# W&B Table
run = wandb.init(project="hpml-hw2-llm", name="c5-summary")

columns = ["batch_size", "lr", "final_train_loss", "final_train_acc", "final_test_acc", "total_time (s)"]
table = wandb.Table(columns=columns)
for r in c5_results:
    table.add_data(
        r["batch_size"], r["lr"],
        round(r["final_train_loss"], 4),
        round(r["final_train_acc"], 4),
        round(r["final_test_acc"], 4),
        round(r["total_time"], 1),
    )

wandb.log({"c5_hyperparam_comparison": table})
wandb.finish()
```
